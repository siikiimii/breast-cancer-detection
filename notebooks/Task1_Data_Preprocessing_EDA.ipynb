# Task 1: Data Preprocessing and Exploratory Analysis

This notebook prepares the **Wisconsin Diagnostic Breast Cancer (WDBC)** dataset for modeling. Steps include:
- Loading the dataset
- Cleaning and encoding
- Standardizing features
- Exploratory Data Analysis (EDA)
- Saving processed data for later use


## 1. Load the Dataset

```
import pandas as pd
# Load dataset
df = pd.read_csv('data.csv')
print(df.shape)
df.head()
```

## 2. Check Missing Values and Data Types

```
print(df.isnull().sum())
print(df.dtypes)
```

## 3. Drop Unnecessary Columns and Encode Labels

```
# Drop columns
if 'id' in df.columns:
    df.drop(columns=['id'], inplace=True)
if 'Unnamed: 32' in df.columns:
    df.drop(columns=['Unnamed: 32'], inplace=True)

# Encode diagnosis
if 'diagnosis' in df.columns:
    df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

## 4. Split Features and Target

```
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

## 5. Standardize Features

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 6. Exploratory Data Analysis

```
# Class distribution
print(y.value_counts())

# Correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Feature distribution
df.hist(figsize=(20, 15))
plt.show()
```

## 7. Verify Processed Data

```
print('Missing values:
', df.isnull().sum())
print('Feature shape:', X_scaled.shape)
print('Target distribution:
', y.value_counts())
print('Scaled feature mean:', X_scaled.mean(), 'std:', X_scaled.std())
```

## 8. Save Processed Data

```
import numpy as np
# Save as NumPy arrays
np.save('X_scaled.npy', X_scaled)
np.save('y.npy', y.to_numpy())

# Save as CSV
processed_df = pd.DataFrame(X_scaled, columns=X.columns)
processed_df['diagnosis'] = y
processed_df.to_csv('processed_data.csv', index=False)
print('Files saved: X_scaled.npy, y.npy, processed_data.csv')
```
